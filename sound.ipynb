```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 6
```

```python
# Harmonics
f, ax = plt.subplots(N, sharex=True, sharey=True)
ax[0].set_title('Harmonics')

for n in range(N):
    x = np.linspace(0, 1, 1000)
    ax[n].plot(x, 0*x, 'b-')
    
    def lam(n): return 2/(n + 1)
    ax[n].plot(x, +np.sin(2*np.pi/lam(n)*x), 'g-')
    ax[n].plot(x, -np.sin(2*np.pi/lam(n)*x), 'g--')
  
    ax[n].axis('off')

f.subplots_adjust(hspace=0)
plt.axis('off')
plt.show()
```

Harmonic wavelengths

\begin{gather}
\lambda_0 = 2 L \qquad
\lambda_n = \frac{2 L}{n + 1}
\end{gather}

Harmonic frequencies

\begin{gather}
f_0 = \frac{c}{\lambda_0} \qquad
f_n = (n + 1) f_0
\end{gather}

All of the frequencies are octives of the base, so they're the same note and it's not that interesting.

```python
# Plucked Harmonics
f, ax = plt.subplots(N, sharex=True, sharey=True)
ax[0].set_title('Plucked Harmonics')

for n in range(N):
    x = np.linspace(0, 1, 1000)
    ax[n].plot(x,  0*x, 'b-')
    
    def lam(n): return 2/(n + 1)
    ax[n].plot(x, +np.sin(2*np.pi/lam(n)*x), 'g-')
    ax[n].plot(x, -np.sin(2*np.pi/lam(n)*x), 'g--')
  
    ax[n].axis('off')

for n in range(1, N): 
    x = np.linspace(0, n/(n + 1), 1000)
    
    def lam(n): return 2*n/(n + 1)
    ax[n].plot(x, +np.sin(2*np.pi/lam(n)*x), 'r-')
    ax[n].plot(x, -np.sin(2*np.pi/lam(n)*x), 'r--')
  
    ax[n].axis('off')

f.subplots_adjust(hspace=0)
plt.axis('off')
plt.show()
```

Plucked harmonic wavelengths and frequencies.

\begin{gather}
\lambda_n' = \frac{2 n L}{n + 1} \qquad
f_n' = \frac{n + 1}{n} f_0
\end{gather}
