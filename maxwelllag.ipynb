### Appendix ###

Let us verify the below equations with some admittedly hackey notation.

First note that

\begin{gather}
[E]_i = c F^{i 0} \qquad
[B]_i = -\tfrac{1}{2} \epsilon_{i j k} F^{j k} \qquad
[D]_i = \tfrac{1}{c} P^{i 0} \qquad
[H]_i = -\tfrac{1}{2} \epsilon_{i j k} P^{j k}
\end{gather}

and show that

\begin{gather}
[E]_i
= c F^{i 0}
= c (\partial^i A^0 - \partial^0 A^i)
= -c \partial_i A^0 - c \partial_0 A^i
= [-\nabla \phi - \partial_t A]_i
\end{gather}

\begin{gather}
[B]_i
= -\tfrac{1}{2} \epsilon_{i j k} F^{j k}
= -\tfrac{1}{2} \epsilon_{i j k} (\partial^j A^k - \partial^k A^j)\
= \epsilon_{i j k} \partial_j A^k
= [\nabla \times A]_i
\end{gather}

\begin{gather}
[\nabla \cdot D]
= \partial_i \tfrac{1}{c} P^{i 0}
= \tfrac{1}{c} \partial_\alpha P^{\alpha 0}
= \tfrac{1}{c} J^0
= [\rho]
\end{gather}

\begin{gather}
[-\partial_t D + \nabla \times H]_i
= -\partial_0 P^{i 0} - \epsilon^{i j k} \partial_j (\tfrac{1}{2} \epsilon_{k m n} P^{m n})
= -\partial_0 P^{i 0} - \partial_j P^{i j}
= \partial_\alpha P^{\alpha i}
= J^i = [J]_i
\end{gather}


Second note that

\begin{gather}
\frac{\partial F^{\alpha \beta}}{\partial [E]_i}
= \frac{1}{c} (\delta^\alpha_i \delta^\beta_0 - \delta^\alpha_0 \delta^\beta_i)
\qquad
\frac{\partial F^{j k}}{\partial [B]_i}
= -\epsilon^{i j k}
\qquad
\frac{\partial A^i}{\partial [\phi]} = \frac{1}{c} \delta^i_0
\qquad
\frac{\partial A^j}{\partial [A]_i} = \delta^j_i
\end{gather}

and show that

\begin{gather}
\frac{\partial \mathcal{L}}{\partial [E]_i}
= \frac{\partial \mathcal{L}}{\partial F^{\alpha \beta}}
\frac{1}{c} (\delta^\alpha_i \delta^\beta_0 - \delta^\alpha_0 \delta^\beta_i)
= \frac{2}{c} \frac{\partial \mathcal{L}}{\partial F^{i 0}}
= -\frac{2}{c} \frac{\partial \mathcal{L}}{\partial F_{i 0}}
= \frac{1}{c} P^{i 0}
= [D]_i
\\
\frac{\partial \mathcal{L}}{\partial [B]_i}
= -\frac{\partial \mathcal{L}}{\partial F^{j k}} \epsilon^{i j k}
= -\frac{\partial \mathcal{L}}{\partial F_{j k}} \epsilon^{i j k}
= \frac{1}{2} \epsilon_{i j k} P^{j k}
= -[H]_i
\\
\frac{\partial \mathcal{L}}{\partial [\phi]}
= \frac{\partial \mathcal{L}}{\partial A^i} \frac{1}{c} \delta^i_0
= \frac{1}{c} \frac{\partial \mathcal{L}}{\partial A^0}
= -\frac{1}{c} J^0
= -[\rho]
\\
\frac{\partial \mathcal{L}}{\partial [A]_i}
= \frac{\partial \mathcal{L}}{\partial A^j} \delta^i_j
= \frac{\partial \mathcal{L}}{\partial A^i}
= -\frac{\partial \mathcal{L}}{\partial A_i}
= J^i
= [J]_i
\end{gather}

### Maxwell's Equations ###

Let us define the following four-vectors and tensors.

\begin{gather}
\partial_\alpha = \begin{pmatrix} \frac{1}{c}\partial_t & \partial_x & \partial_y & \partial_z \end{pmatrix}
\qquad
A^\alpha = \left( \begin{matrix}\frac{1}{c}\phi & A_x & A_y & A_z \end{matrix}\right)
\qquad
J^\alpha = \begin{pmatrix} c\rho & J_x & J_y & J_z \end{pmatrix}  
\end{gather}

\begin{gather}
F^{\alpha \beta} = \begin{pmatrix}
0       & -E_x / c & -E_y / c & -E_z / c \\
E_x / c &  0       & -B_z     &  B_y     \\
E_y / c &  B_z     &    0     & -B_x     \\
E_z / c & -B_y     &  B_x     &  0       \end{pmatrix}
\qquad
P^{\alpha\beta} = \begin{pmatrix}
0 & -c D_x & -c D_y & -c D_z \\
 c D_x & 0 & -H_z & H_y \\
 c D_y & H_z & 0 & -H_x \\
 c D_z & -H_y & H_x & 0
\end{pmatrix}
\end{gather}

With these, Maxwell's equations can be written in a compact form.

\begin{gather}
F^{\alpha \beta} = \partial^\alpha A^\beta - \partial^\beta A^\alpha
\quad \Rightarrow \quad
\begin{cases}
E = -\nabla \phi - \dfrac{\partial A}{\partial t} \\
B = \nabla \times A \vphantom{\dfrac{\partial A}{\partial t}}
\end{cases}
\end{gather}

The first pair of equations can be viewed as the definition of the electric and magnetic fields.  As we will see, the second pair of equations can be derived from a general Lagrangian. 

In general, we have $\mathcal{L}(\partial_\alpha A_\beta, A_\beta)$.  This notation is a different.  The Lagrangian in not a function of $\partial_\alpha A_\beta$ and $A_\beta$ for particular values of $\alpha$ and $\beta$.  Rather it is a functions of all combinations of values between 0 and 3. That is, the Lagrangian is a function of twenty variables.  Minimumizing the action gives

\begin{gather}
\partial_\alpha \frac{\partial \mathcal{L}}{\partial (\partial_\alpha A_\beta)}
= \frac{\partial \mathcal{L}}{\partial A_\beta}
\end{gather}

For the particular case where $\mathcal{L}(F_{\alpha\beta}, A_\beta)$, one can show that the previuos equation becomes

\begin{gather}
\partial_\alpha 2 \frac{\partial \mathcal{L}}{\partial F_{\alpha\beta}}
= \frac{\partial \mathcal{L}}{\partial A_\beta}
\end{gather}

With the following definitons
\begin{gather}
P^{\alpha\beta} = -2 \frac{\partial \mathcal{L}}{\partial F_{\alpha\beta}}
\quad \Rightarrow \quad
\begin{cases}
D = \nabla_E \mathcal{L} \vphantom{\dfrac{\partial A}{\partial t}} \\
H = -\nabla_B \mathcal{L} \vphantom{\dfrac{\partial A}{\partial t}}
\end{cases}
\end{gather}

\begin{gather}
J^\beta = -\frac{\partial \mathcal{L}}{\partial A_\beta}
\quad \Rightarrow \quad
\begin{cases}
\rho = - \dfrac{\partial \mathcal{L}}{\partial \phi} \vphantom{\dfrac{\partial A}{\partial t}} \\
J = \nabla_A \mathcal{L} \vphantom{\dfrac{\partial A}{\partial t}}
\end{cases}
\end{gather}

the previous equation becomes

\begin{gather}
\partial_\alpha P^{\alpha \beta} = J^\beta
\quad \Rightarrow \quad
\begin{cases}
\nabla \cdot D = \rho \vphantom{\dfrac{\partial A}{\partial t}} \\
- \dfrac{\partial D}{\partial t} + \nabla \times H = J
\end{cases}
\end{gather}

We see that the second pair of Maxwell's equations can be seen as arising from a general Lagrangian.  In general, $D$ and $H$ are complicated functions of $E$ and $B$.

### The General Energy-Momentum Tensor ###

The following equation is derived in an AJP paper using Noether's theorem.  As before, we assume $\mathcal{L}(A_\alpha, F_{\alpha\beta})$ and use our previous definitions.

\begin{gather}
\Theta^{\alpha}_{\phantom\alpha \beta}
= -J^\alpha A_\beta
- P^{\alpha \gamma} F_{\beta \gamma} - \delta^{\alpha}_{\phantom\alpha \beta} \mathcal{L}
\end{gather}

\begin{split}
\Theta^{0 0}
&= \Theta^{0}_{\phantom\alpha 0} \\
&= -J^0 A_0 - P^{0 \gamma} F_{0 \gamma} - \delta^{0}_{\phantom\alpha 0} \mathcal{L} \\
&= -J^0 A_0 - P^{0 0} F_{0 0} - P^{0 i} F_{0 i} - \mathcal{L} \\
&= -J^0 A^0 + P^{i 0} F^{i 0} - \mathcal{L} \\
&= -\rho \phi + D \cdot E - \mathcal{L} \\
\end{split}

\begin{split}
\Theta^{0 i}
&= -\Theta^{0}_{\phantom\alpha i} \\
&= J^0 A_i + P^{0 \gamma} F_{i \gamma} + \delta^0_{\phantom\alpha i} \mathcal{L} \\
&= J^0 A_i + P^{0 0} F_{i 0} + P^{0 j} F_{i j} \\
&= -J^0 A^i - P^{j 0} F^{i j} \\
&= -[c \rho] [A]_i + \epsilon^{i j k}  c [D]_j [B]_k \qquad \\
&= [-c \rho A + c D \times B]_i
\end{split}

\begin{split}
\Theta^{i 0}
&= -\Theta^{i}_{\phantom\alpha 0} \\
&= J^i A_0 + P^{i \gamma} F_{0 \gamma} + \delta^i_{\phantom\alpha 0} \mathcal{L} \\
&= J^i A_0 + P^{i 0} F_{0 0} + P^{i j} F_{0 j} \\
&= J^i A^0 - P^{i j} F^{0 j} \\
\end{split}

\begin{gather}
F^{i 0} = \frac{1}{c} [E]_i \qquad
F^{i j} = -\epsilon^{i j k} [B]_k \qquad
P^{i 0} = c [D]_i \qquad
P^{i j} = -\epsilon^{i j k} [H]_k \qquad
\end{gather}
