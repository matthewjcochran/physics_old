### Maxwell's Equations ###

Let us define the following four-vectors and tensors.

\begin{gather}
\partial_\alpha = \begin{pmatrix} \frac{1}{c}\partial_t & \partial_x & \partial_y & \partial_z \end{pmatrix}
\qquad
A^\alpha = \left( \begin{matrix}\frac{1}{c}\phi & A_x & A_y & A_z \end{matrix}\right)
\qquad
J^\alpha = \begin{pmatrix} c\rho & J_x & J_y & J_z \end{pmatrix}  
\end{gather}

\begin{gather}
F^{\alpha \beta} = \begin{pmatrix}
0  & -\tfrac{1}{c} E_x & -\tfrac{1}{c} E_y & -\tfrac{1}{c} E_z \\
\tfrac{1}{c} E_x &  0   & -B_z &  B_y     \\
\tfrac{1}{c} E_y &  B_z &    0 & -B_x     \\
\tfrac{1}{c} E_z & -B_y &  B_x &  0 \end{pmatrix}
\qquad
P^{\alpha\beta} = \begin{pmatrix}
0 & -c D_x & -c D_y & -c D_z \vphantom{\tfrac{1}{c}} \\
 c D_x & 0 & -H_z & H_y \vphantom{\tfrac{1}{c}} \\
 c D_y & H_z & 0 & -H_x \vphantom{\tfrac{1}{c}} \\
 c D_z & -H_y & H_x & 0 \vphantom{\tfrac{1}{c}}
\end{pmatrix}
\end{gather}

With these, Maxwell's equations can be written in a compact form.

\begin{gather}
F^{\alpha \beta} = \partial^\alpha A^\beta - \partial^\beta A^\alpha
\quad \Rightarrow \quad
\begin{cases}
E = -\nabla \phi - \dfrac{\partial A}{\partial t} \\
B = \nabla \times A \vphantom{\dfrac{\partial A}{\partial t}}
\end{cases}
\end{gather}

The first pair of equations can be viewed as the definition of the electric and magnetic fields.  As we will see, the second pair of equations can be derived from a general Lagrangian. 

In general, we have $\mathcal{L}(\partial_\alpha A_\beta, A_\beta)$.  This notation is a different.  The Lagrangian in not a function of $\partial_\alpha A_\beta$ and $A_\beta$ for particular values of $\alpha$ and $\beta$.  Rather it is a functions of all combinations of values between 0 and 3. That is, the Lagrangian is a function of twenty variables.  Minimumizing the action gives

\begin{gather}
\partial_\alpha \frac{\partial \mathcal{L}}{\partial (\partial_\alpha A_\beta)}
= \frac{\partial \mathcal{L}}{\partial A_\beta}
\end{gather}

For the particular case where $\mathcal{L}(F_{\alpha\beta}, A_\beta)$, one can show that the previuos equation becomes

\begin{gather}
\partial_\alpha 2 \frac{\partial \mathcal{L}}{\partial F_{\alpha\beta}}
= \frac{\partial \mathcal{L}}{\partial A_\beta}
\end{gather}

With the following definitons
\begin{gather}
P^{\alpha\beta} = -2 \frac{\partial \mathcal{L}}{\partial F_{\alpha\beta}}
\quad \Rightarrow \quad
\begin{cases}
D = \nabla_E \mathcal{L} \vphantom{\dfrac{\partial A}{\partial t}} \\
H = -\nabla_B \mathcal{L} \vphantom{\dfrac{\partial A}{\partial t}}
\end{cases}
\end{gather}

\begin{gather}
J^\beta = -\frac{\partial \mathcal{L}}{\partial A_\beta}
\quad \Rightarrow \quad
\begin{cases}
\rho = - \dfrac{\partial \mathcal{L}}{\partial \phi} \vphantom{\dfrac{\partial A}{\partial t}} \\
J = \nabla_A \mathcal{L} \vphantom{\dfrac{\partial A}{\partial t}}
\end{cases}
\end{gather}

the previous equation becomes

\begin{gather}
\partial_\alpha P^{\alpha \beta} = J^\beta
\quad \Rightarrow \quad
\begin{cases}
\nabla \cdot D = \rho \vphantom{\dfrac{\partial A}{\partial t}} \\
- \dfrac{\partial D}{\partial t} + \nabla \times H = J
\end{cases}
\end{gather}

We see that the second pair of Maxwell's equations can be seen as arising from a general Lagrangian.  In general, $D$ and $H$ are complicated functions of $E$ and $B$.

&= \int \left[ \frac{\partial \mathcal L}{\partial A_\beta} \delta A_\beta 
+ \frac{\partial \mathcal L}{\partial (\partial_\alpha A_\beta)} \delta (\partial_\alpha A_\beta) \right] dx^4 \\
&= \int \left[ \frac{\partial \mathcal L}{\partial A_\beta} \delta A_\beta 
+ \frac{\partial \mathcal L}{\partial (\partial_\alpha A_\beta)} \partial_\alpha (\delta A_\beta)
\right] dx^4 \\
&= \int \left[
\left( \frac{\partial \mathcal L}{\partial A_\beta}
- \partial_\alpha \frac{\partial \mathcal L}{\partial (\partial_\alpha A_\beta)} \right) \delta A_\beta
+ \partial_\alpha \left( \frac{\partial \mathcal L}{\partial (\partial_\alpha A_\beta)} \delta A_\beta \right)
\right] dx^4
\end{split}

\begin{split}
\delta S
&= \int \mathcal \delta L(A_\beta, F_{\alpha \beta}) dx^4 \\
&= \int \left[ \frac{\partial \mathcal L}{\partial A_\beta} \delta A_\beta 
+ \frac{\partial \mathcal L}{\partial  F_{\alpha \beta}}
\delta  F_{\alpha \beta} \right] dx^4 \\
&= \int \left[ \frac{\partial \mathcal L}{\partial A_\beta} \delta A_\beta 
+ \frac{\partial \mathcal L}{\partial  F_{\alpha \beta}}
\delta ( \partial_\alpha A_\beta - \partial_\beta A_\alpha ) \right] dx^4 \\
&= \int \left[ \frac{\partial \mathcal L}{\partial A_\beta} \delta A_\beta 
+ \frac{\partial \mathcal L}{\partial  F_{\alpha \beta}}
\partial_\alpha (\delta A_\beta)
- \frac{\partial \mathcal L}{\partial  F_{\alpha \beta}}
\partial_\beta (\delta A_\alpha ) \right] dx^4 \\
&= \int \left[ \frac{\partial \mathcal L}{\partial A_\beta} \delta A_\beta 
+ 2 \frac{\partial \mathcal L}{\partial  F_{\alpha \beta}}
\partial_\alpha (\delta A_\beta) \right] dx^4 \\
&= \int \left[ \frac{\partial \mathcal L}{\partial A_\beta} \delta A_\beta 
+ \partial_\alpha \left( 2 \frac{\partial \mathcal L}{\partial  F_{\alpha \beta}}
\delta A_\beta \right)
- \left( \partial_\alpha 2 \frac{\partial \mathcal L}{\partial  F_{\alpha \beta}} \right)
\delta A_\beta \right] dx^4 \\
&= \int \left[ \left( \frac{\partial \mathcal L}{\partial A_\beta}
- \partial_\alpha 2 \frac{\partial \mathcal L}{\partial  F_{\alpha \beta}} \right)
\delta A_\beta
+ \partial_\alpha \left( 2 \frac{\partial \mathcal L}{\partial  F_{\alpha \beta}}
\delta A_\beta \right)
 \right] dx^4 \\
\end{split}

\begin{split}
\delta S
&= \int \mathcal \delta
\left( -\dfrac{1}{2} P^{\alpha \beta} F_{\alpha \beta} - \mathcal H(P_{\alpha \beta}, A_\beta) \right) dx^4 \\
&= -\int \left(
\dfrac{1}{2} F^{\alpha \beta} \delta P_{\alpha \beta} 
+ \dfrac{1}{2} P^{\alpha \beta} \delta F_{\alpha \beta}
+ \frac{\partial \mathcal H}{\partial P_{\alpha \beta}} \delta P_{\alpha \beta}
+ \frac{\partial \mathcal H}{\partial A_\beta} \delta A_\beta
\right) dx^4 \\
&= -\int \left[
\left( \dfrac{1}{2} F^{\alpha \beta}
+ \frac{\partial \mathcal H}{\partial P_{\alpha \beta}} \right) \delta P_{\alpha \beta}
+ \frac{\partial \mathcal H}{\partial A_\beta} \delta A_\beta
+ \dfrac{1}{2} P^{\alpha \beta} \delta (\partial_\alpha A_\beta - \partial_\beta A_\alpha)
\right] dx^4 \\
&= -\int \left[
\left( \dfrac{1}{2} F^{\alpha \beta}
+ \frac{\partial \mathcal H}{\partial P_{\alpha \beta}} \right) \delta P_{\alpha \beta}
+ \frac{\partial \mathcal H}{\partial A_\beta} \delta A_\beta
+ \dfrac{1}{2} P^{\alpha \beta} \partial_\alpha \delta A_\beta
- \dfrac{1}{2} P^{\alpha \beta} \partial_\beta \delta A_\alpha
\right] dx^4 \\
&= -\int \left[
\left( \dfrac{1}{2} F^{\alpha \beta}
+ \frac{\partial \mathcal H}{\partial P_{\alpha \beta}} \right) \delta P_{\alpha \beta}
+ \frac{\partial \mathcal H}{\partial A_\beta} \delta A_\beta
+ P^{\alpha \beta} \partial_\alpha \delta A_\beta
\right] dx^4 \\
&= -\int \left[
\left( \dfrac{1}{2} F^{\alpha \beta}
+ \frac{\partial \mathcal H}{\partial P_{\alpha \beta}} \right) \delta P_{\alpha \beta}
+ \frac{\partial \mathcal H}{\partial A_\beta} \delta A_\beta
+ \partial_\alpha (P^{\alpha \beta} \delta A_\beta)
- (\partial_\alpha P^{\alpha \beta}) \delta A_\beta
\right] dx^4 \\
&= -\int \left[
\left( \dfrac{1}{2} F^{\alpha \beta}
+ \frac{\partial \mathcal H}{\partial P_{\alpha \beta}} \right) \delta P_{\alpha \beta}
+ \left( \frac{\partial \mathcal H}{\partial A_\beta} 
- \partial_\alpha P^{\alpha \beta} \right) \delta A_\beta
+ \partial_\alpha (P^{\alpha \beta} \delta A_\beta)
\right] dx^4 \\
\end{split}

Let us consider the Hamiltonian.

\begin{split}
\mathcal H
&= \dfrac{\partial \mathcal L}{\partial F_{\alpha \beta}} F^{\alpha \beta} - \mathcal L \\
&= -\dfrac{1}{2} P^{\alpha \beta} F_{\alpha \beta} - \mathcal L \\
&= D \cdot E - \mathcal L \vphantom{\dfrac{1}{2}} \\
\end{split}

We see that it is the same as the energy density, $u$.  One can show the following

\begin{gather}
F^{\alpha\beta} = -2 \frac{\partial \mathcal H}{\partial P_{\alpha\beta}}
\quad \Rightarrow \quad
\begin{cases}
E = \nabla_D \mathcal H \vphantom{\dfrac{\partial A}{\partial t}} \\
B = \nabla_H \mathcal H \vphantom{\dfrac{\partial A}{\partial t}}
\end{cases}
\end{gather}

\begin{gather}
\partial_\alpha P^{\alpha \beta} = \frac{\partial \mathcal H}{\partial A_\beta}
\quad \Rightarrow \quad
\begin{cases}
\nabla \cdot D = \dfrac{\partial \mathcal H}{\partial \phi} \\
- \dfrac{\partial D}{\partial t} + \nabla \times H = -\nabla_A \mathcal H
\end{cases}
\end{gather}
