### Fluid mechanics

Conservation of mass.

\begin{gather}
\frac{\partial \rho}{\partial t}
+ \frac{\partial (\rho u)}{\partial x}
+ \frac{\partial (\rho v)}{\partial y}
+ \frac{\partial (\rho w)}{\partial z}
= 0
\end{gather}

Conservation of momentum.

\begin{gather}
\frac{\partial (\rho u)}{\partial t}
+ \frac{\partial (\rho u^2)}{\partial x}
+ \frac{\partial (\rho v u)}{\partial y}
+ \frac{\partial (\rho w u)}{\partial z}
= f_x \\
\frac{\partial (\rho v)}{\partial t}
+ \frac{\partial (\rho u v)}{\partial x}
+ \frac{\partial (\rho v^2)}{\partial y}
+ \frac{\partial (\rho w v)}{\partial z}
= f_y \\
\frac{\partial (\rho w)}{\partial t}
+ \frac{\partial (\rho u w)}{\partial x}
+ \frac{\partial (\rho v w)}{\partial y}
+ \frac{\partial (\rho w^2)}{\partial z}
= f_z
\end{gather}

### Math

Use the chain rule.  Using conservation of mass, the momentum equations can be rewritten.

\begin{gather}
\rho \left( \frac{\partial u}{\partial t}
+ u \frac{\partial u}{\partial x}
+ v \frac{\partial u}{\partial y}
+ w \frac{\partial u}{\partial z}
\right) = f_x \\
\rho \left( \frac{\partial v}{\partial t}
+ u \frac{\partial v}{\partial x}
+ v \frac{\partial v}{\partial y}
+ w \frac{\partial v}{\partial z}
\right) = f_y \\
\rho \left( \frac{\partial w}{\partial t}
+ u \frac{\partial w}{\partial x}
+ v \frac{\partial w}{\partial y}
+ w \frac{\partial w}{\partial z}
\right) = f_z
\end{gather}

We can rewrite the momentum equations with the material derivative.

\begin{gather}
\rho \frac{D u}{D t} = f_x \qquad
\rho \frac{D v}{D t} = f_y \qquad
\rho \frac{D w}{D t} = f_z
\end{gather}

### Pressure

\begin{gather}
f_x = - \frac{\partial p}{\partial x} \qquad
f_y = - \frac{\partial p}{\partial y} \qquad
f_z = - \frac{\partial p}{\partial z}
\end{gather}

### More math

Let us go back to conservations of mass.

\begin{gather}
\frac{D \rho}{D t}
+ \rho \left( \frac{\partial u}{\partial x}
+ \frac{\partial v}{\partial y}
+ \frac{\partial w}{\partial z} \right)
= 0
\end{gather}

Let us revisit the momentum equations.

\begin{gather}
\frac{D (\rho u)}{D t} + \rho u \left(
\frac{\partial u}{\partial x} + \frac{\partial v}{\partial y} + \frac{\partial w}{\partial z}
\right) = f_x \\
\frac{D (\rho v)}{D t} + \rho v \left(
\frac{\partial u}{\partial x} + \frac{\partial v}{\partial y} + \frac{\partial w}{\partial z}
\right) = f_y \\
\frac{D (\rho w)}{D t} + \rho w \left(
\frac{\partial u}{\partial x} + \frac{\partial v}{\partial y} + \frac{\partial w}{\partial z}
\right) = f_z
\end{gather}

What is the physical interpretation of the material derivative?  Looks like a kinematical type quantum potential.

### Vectors

Lets us go to vector notation, finally.

\begin{gather}
\frac{\partial \rho}{\partial t} + \nabla \cdot (\rho \mathbf{u}) = 0 \\
\rho \left( \frac{\partial \mathbf{u}}{\partial t}
+ (\mathbf{u} \cdot \nabla) \mathbf{u}
\right) = \mathbf{f}
\end{gather}

### Vector Zero

\begin{gather}
\nabla \times (\nabla \phi)
= \left( \frac{\partial}{\partial y} \frac{\partial \phi}{\partial z}
       - \frac{\partial}{\partial z} \frac{\partial \phi}{\partial y} \right) \hat x
+ \left( \frac{\partial}{\partial z} \frac{\partial \phi}{\partial x}
       - \frac{\partial}{\partial x} \frac{\partial \phi}{\partial z} \right) \hat y
+ \left( \frac{\partial}{\partial x} \frac{\partial \phi}{\partial y}
       - \frac{\partial}{\partial y} \frac{\partial \phi}{\partial x} \right) \hat z
= 0
\end{gather}

\begin{gather}
\nabla \cdot (\nabla \times \mathbf u)
= \frac{\partial}{\partial x} \left( \frac{\partial w}{\partial y} - \frac{\partial v}{\partial z} \right)
+ \frac{\partial}{\partial y} \left( \frac{\partial u}{\partial z} - \frac{\partial w}{\partial x} \right)
+ \frac{\partial}{\partial z} \left( \frac{\partial v}{\partial x} - \frac{\partial u}{\partial y} \right)
=  0
\end{gather}

### Levi-Civita

\begin{gather}
[\mathbf{a} \times \mathbf{b}]_k = \epsilon_{i j k} a_i b_j
\end{gather}

\begin{gather}
\epsilon_{i j k} \epsilon_{m n k} = \delta_{i m}\delta_{j n} - \delta_{i n}\delta_{j m}
\end{gather}

\begin{split}
[\mathbf a \times (\mathbf b \times \mathbf c)]_k
&= \epsilon_{i j k} a_i (\epsilon_{m n j} b_m c_n) \\
&= \epsilon_{k i j} \epsilon_{m n j} a_i b_m c_n \\
&= (\delta_{k m}\delta_{i n} - \delta_{k n}\delta_{i m}) a_i b_m c_n \\
&= \delta_{k m}\delta_{i n} a_i b_m c_n - \delta_{k n}\delta_{i m} a_i b_m c_n \\
&= a_i b_k c_i - a_i b_i c_k \\ 
&= [\mathbf b (\mathbf a \cdot \mathbf c) - \mathbf c (\mathbf a \cdot \mathbf b)]_k \\ 
\end{split}

\begin{split}
[\nabla \times (\nabla \times \mathbf a)]_k
&= \epsilon_{i j k} \partial_i (\epsilon_{m n j} \partial_m a_n) \\
&= \epsilon_{k i j} \epsilon_{m n j} \partial_i \partial_m a_n \\
&= (\delta_{k m}\delta_{i n} - \delta_{k n}\delta_{i m}) \partial_i \partial_m a_n \\
&= \delta_{k m}\delta_{i n} a_i b_m c_n - \delta_{k n}\delta_{i m} \partial_i \partial_m a_n \\
&= \partial_i \partial_k a_i - \partial_i \partial_i a_k \\
&= [\nabla (\nabla \cdot \mathbf a) - \nabla^2 \mathbf a]_k
\end{split}

\begin{split}
[\mathbf a \times (\nabla \times \mathbf a)]_k
&= \epsilon_{i j k} a_i (\epsilon_{m n j} \partial_m a_n) \\
&= \epsilon_{k i j} \epsilon_{m n j} a_i \partial_m a_n \\
&= (\delta_{k m}\delta_{i n} - \delta_{k n}\delta_{i m}) a_i \partial_m a_n \\
&= \delta_{k m}\delta_{i n} a_i \partial_m a_n - \delta_{k n}\delta_{i m} a_i \partial_m a_n \\
&= a_i \partial_k a_i - a_i \partial_i a_k \\
&= \left[ \nabla \tfrac{1}{2} \mathbf a^2 - (\mathbf a \cdot \nabla) \mathbf a \right]_k \\
\end{split}


\begin{gather}
\nabla \times[\mathbf a \times (\nabla \times \mathbf a)]
= - \nabla \times [(\mathbf a \cdot \nabla) \mathbf a]
\end{gather}

### Vertex Speculation

Assume that density is constant.

\begin{gather}
\nabla \cdot \mathbf u = 0 \\
\frac{\partial \mathbf{u}}{\partial t}
+ (\mathbf u \cdot \nabla) \mathbf u = 0
\end{gather}

Introduce vorticity.

\begin{gather}
\mathbf w = \nabla \times \mathbf u
\end{gather}

\begin{gather}
\frac{\partial}{\partial t} \nabla \times \mathbf u
+ \nabla \times [(\mathbf u \cdot \nabla) \mathbf u] = 0
\end{gather}

\begin{gather}
\frac{\partial}{\partial t} \nabla \times \mathbf u
- \nabla \times[\mathbf u \times (\nabla \times \mathbf u)] = 0
\end{gather}

\begin{gather}
\frac{\partial \mathbf w}{\partial t}
- \nabla \times (\mathbf u \times \mathbf w) = 0
\end{gather}
