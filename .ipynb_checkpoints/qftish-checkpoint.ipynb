Let us consider the vector potential in one dimension in the absence of current.  It satisfies the wave equation.

\begin{split}
\frac{\partial^2 A}{\partial x^2} - \frac{1}{c^2} \frac{\partial^2 A}{\partial t^2} = 0
\end{split}

In this case, the Lagrangian density is

\begin{split}
\mathcal{L} = \frac{1}{2 \mu_0} \left[ \left( \partial_t A \right)^2 - c^2 \left( \partial_x A \right)^2 \right]
\end{split}



Let us consider the spatial Fourier transform of $A(x)$ along a line of length $L$.  Defining $k_n = \frac{2 \pi n}{\ell}$, we have

\begin{split}
A(x, t) = \frac{1}{\ell} \sum_{n = -\infty}^\infty a_n(t) e^{i k_n x}
\end{split}


\begin{split}
a_n(t) = \int_{0}^{\ell} A(x, t) e^{-i k_n x} dx
\end{split}

In general, $a_n$ is comples.  Since, however, $A(x)$ is real we have $a_n^* = a_{-n}.$  Also, note that $k_{-n} = - k_n$. Consistancy considerations lead to the delta function equations.

\begin{split}
\delta(x, x') = \frac{1}{\ell} \sum_{n = -\infty}^\infty e^{i k_n (x - x')}
\end{split}

\begin{split}
\delta_{n, n'} = \frac{1}{\ell} \int_{0}^{L} e^{-i (k_n - k_{n'}) x} dx
\end{split}

Using the above, let us find the actual Lagrangian.

\begin{split}
L
&= \int_0^\ell \mathcal{L} dx \\
&= \int_0^\ell \frac{1}{2 \mu_0}\left[
\left( \partial_t A \right)
\left( \partial_t A \right)
- c^2
\left( \partial_x A \right)
\left( \partial_x A \right)
\right] dx \\
&= \int_0^\ell \frac{1}{2 \mu_0}\left[
\left( \frac{1}{\ell} \sum_{n = -\infty}^\infty \dot a_n e^{i k_n x} \right)
\left( \frac{1}{\ell} \sum_{n' = -\infty}^\infty \dot a_{n'} e^{i k_{n'} x} \right)
- c^2 
\left( \frac{1}{\ell} \sum_{n = -\infty}^\infty i k_n a_n e^{i k_n x} \right)
\left( \frac{1}{\ell} \sum_{n' = -\infty}^\infty i k_{n'} a_{n'} e^{i k_{n'} x} \right)
\right] dx \\
&= \frac{1}{2 \mu_0} \frac{1}{\ell} \sum_{n = -\infty}^\infty \frac{1}{\ell} \sum_{n' = -\infty}^\infty
\int_0^\ell \left(
\dot a_n \dot a_{n'} + c^2 k_n k_{n'} a_n a_{n'}
\right) e^{i (k_n + k_{n'}) x} dx \\
&= \frac{1}{2 \mu_0} \frac{1}{\ell} \sum_{n = -\infty}^\infty \sum_{n' = -\infty}^\infty
\left( \frac{1}{\ell} \int_0^\ell e^{i (k_n + k_{n'}) x} dx \right) \left(
\dot a_n \dot a_{n'} + c^2 k_n k_{n'} a_n a_{n'}
\right) \\
&= \frac{1}{2 \mu_0} \frac{1}{\ell} \sum_{n = -\infty}^\infty \sum_{n' = -\infty}^\infty
\delta_{-n, n'} \left( \dot a_n \dot a_{n'} + c^2 k_n k_{n'} a_n a_{n'} \right) \\
&= \frac{1}{2 \mu_0} \frac{1}{\ell} \sum_{n = -\infty}^\infty
\left( \dot a_n \dot a^*_n - c^2 k_n^2 a_n a^*_n \right) \\
&= \frac{1}{\ell} \sum_{n = -\infty}^\infty \frac{1}{2 \mu_0}
\left( \dot a^*_n \dot a_n - \omega_n^2 a^*_n a_n \right) \\
\end{split}

Let us try to interpret the $a_n$ in terms of real quantites.  Let us say $a^*_n = q_n + i q'_n$.  Let us see what we can learn from this change in notation.

\begin{split}
L &= \frac{1}{\ell} \sum_{n = -\infty}^\infty \frac{1}{2 \mu_0} \left[
\left( \dot q^2_n - \omega_n^2 q^2_n \right)
+ \left( \dot q'^2_n - \omega_n^2 q'^2_n \right) \right] \\
\end{split}

It appears that we have two oscillators, one for the sine term and one for the cosine term.

\begin{split}
A(x)
&= \frac{1}{\ell} \sum_{n = -\infty}^\infty \left[
q_n \cos(k_n x) + q'_n \sin(k_n x) \right]
\end{split}


\begin{split}
q_n  &= \int_{0}^{\ell} A(x) \cos(k_n x) dx \\
q'_n &= \int_{0}^{\ell} A(x) \sin(k_n x) dx \\
\end{split}



Plugging back into the first equations we have the following symmetric relationship.

$$ A(x, y, z, t) = \frac{1}{\ell^3}
            \sum_{n_x} \sum_{n_y} \sum_{n_z}
            a_{n_x n_y n_z}(t) 
            e^{i (k_x x + k_y y + k_z z)} $$

$$ a_{n_x n_y n_z}(t) = \int_0^\ell \int_0^\ell \int_0^\ell
            A(x, y, z, t)
            e^{-i (k_x x + k_y y + k_z z)} dx dy dz $$
            
Notational update.  It is very compact.

$$ A(r, t) = \frac{1}{\ell^3} \sum_n a_n(t) e^{i k_n \cdot r} $$

$$ a_n(t) = \int A(r, t) e^{-i k_n \cdot r} d^3 r $$
            
Let us get to the delta functions.

$$ \delta(r, r') = \frac{1}{\ell^3} \sum_n e^{i k_n \cdot (r - r')} $$

$$ \delta_{n, n'} = \frac{1}{\ell^3} \int e^{-i (k_n - k_{n'}) \cdot r} d^3r $$



<h3>The Coulomb gauge</h3>

\begin{split}
E = -\frac{\partial A}{\partial t} \\
B = \nabla \times A \\
\end{split}

Plugging into the second two equations, we see that they are automatically satisfied.

\begin{split}
\nabla \times E + \frac{\partial B}{\partial t}
= - \nabla \times \frac{\partial A}{\partial t} + \frac{\partial}{\partial t} \nabla \times A = 0 \\
\end{split}

\begin{split}
\nabla \cdot B = \nabla \cdot (\nabla \times A) = 0
\end{split}

Now we can look at what the other two equations tell us.

\begin{gather}
\nabla \cdot E = \frac{\rho}{\epsilon_0} \\
\nabla \times B - \frac{1}{c^2} \frac{\partial E}{\partial t} = \mu_0 J
\end{gather}

Math

\begin{gather}
- \frac{\partial}{\partial t} \nabla \cdot A = \frac{\rho}{\epsilon_0} \\
\nabla (\nabla \cdot A) - \nabla^2 A + \frac{\partial^2 A}{\partial t^2} = \mu_0 J
\end{gather}

Let us rewrite.

\begin{gather}
\frac{\partial}{\partial t} (\nabla \cdot A) = - \frac{\rho}{\epsilon_0} \\
\nabla^2 A - \frac{\partial^2 A}{\partial t^2} - \nabla (\nabla \cdot A) = - \mu_0 J
\end{gather}

Since we must have

\begin{split}
\nabla \cdot A = \frac{1}{\ell^3} \sum_n (k_n \cdot a_n) e^{i k_n \cdot r} = 0
\end{split}

we introduce two vecotors, $\alpha_{1n}$ and $\alpha_{2n}$, which are each normal to $k_n$ and eachother and write

\begin{split}
A = \frac{1}{\ell^3} \sum_n (\alpha_n + \alpha'_n) e^{i k_n \cdot r}
\end{split}

Now define the electric and magnetic fields.

\begin{split}
E = - \frac{1}{\ell^3} \sum_n (\dot \alpha_n + \dot \alpha'_n) e^{i k_n \cdot r}
\end{split}

\begin{split}
B = \frac{1}{\ell^3} \sum_n i k_n \times (\alpha_n + \alpha'_n) e^{i k_n \cdot r}
\end{split}

\begin{split}
\int E^2 dV
&= \frac{1}{\ell^3} \sum_n (\dot \alpha_n + \dot \alpha'_n) \cdot (\dot \alpha^*_n + \dot \alpha'^*_n) \\
&= \frac{1}{\ell^3} \sum_n \left[
(\dot \alpha^*_n \cdot \dot \alpha_n) + (\dot \alpha'^*_n \cdot \dot \alpha'_n) \right]
\end{split}

\begin{split}
\int B^2 dV
&= \frac{1}{\ell^3} \sum_n (k_n \times \alpha_n + k_n \times \alpha'_n) \cdot (k_n \times \alpha^*_n + k_n \times \alpha'^*_n) \\
&= \frac{1}{\ell^3} \sum_n \left[
(k_n \times \alpha^*_n \cdot k_n \times \alpha_n) + (k_n \times \alpha'^*_n \cdot k_n \times \alpha'_n) \right] \\
&= \frac{1}{\ell^3} \sum_n k_n^2 \left[
(\alpha^*_n \cdot \alpha_n) + (\alpha'^*_n \cdot \alpha'_n)
\right]
\end{split}

\begin{split}
L &= \int \frac{1}{2 \mu_0} \left( \frac{E^2}{c^2} - B^2 \right) dV \\
&= \frac{1}{2 c^2 \mu_0}
\frac{1}{\ell^3} \sum_n \left[
(\dot \alpha^*_n \cdot \dot \alpha_n) + (\dot \alpha'^*_n \cdot \dot \alpha'_n)
- c^2 k_n^2 (\alpha^*_n \cdot \alpha_n)
- c^2 k_n^2 (\alpha'^*_n \cdot \alpha'_n)
\right]
\end{split}
