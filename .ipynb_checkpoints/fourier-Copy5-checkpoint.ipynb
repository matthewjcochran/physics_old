### Fourier transform: Discrete 1D ###

Consider a real function $f(x)$ between $0$ and $L$.  We write this as a sum of periodic functions.

#### Base defintion ####

Start with real variables and positive indicies.

\begin{gather}
f(x) = a_0 + \sum_{n = 1}^{\infty}
\left[ a_n \cos \left(\frac{2 \pi n}{L} x\right) + b_n \sin \left(\frac{2 \pi n}{L} x\right) \right]
\end{gather}

Calculations yeild the following.

\begin{gather}
a_0 = \frac{1}{L} \int_0^L f(x) dx \\
a_n = \frac{2}{L} \int_0^L f(x) \cos \left(\frac{2 \pi n}{L} x\right) dx \\
b_n = \frac{2}{L} \int_0^L f(x) \sin \left(\frac{2 \pi n}{L} x\right) dx
\end{gather}

Note that

\begin{gather}
a_{-n} = a_n \qquad
b_{-n} = -b_n
\end{gather}

#### First notation update ####

Set $k_n = 2 \pi n / L$, allow $n$ to have integer values, and introduce $a_n$ and $b_n$.  

\begin{gather}
f(x) = \sum_{n = -\infty}^{\infty} \left[ A_n \cos \left(k_n x\right) + B_n \sin \left(k_n x\right) \right]
\end{gather}

Calculations yeild the following.

\begin{gather}
A_n = \frac{1}{L} \int_0^L f(x) \cos \left(k_n x\right) dx \\
B_n = \frac{1}{L} \int_0^L f(x) \sin \left(k_n x\right) dx
\end{gather}


Note that

\begin{gather}
A_{-n} = A_n \qquad
B_{-n} = -B_n
\end{gather}

Connect this to the old notation.

\begin{gather}
a_0 = A_0 \qquad
a_n = 2 A_n \qquad
b_n = 2 B_n
\end{gather}

#### Second notation update ####

Replace the tigonmetric functions with complex exponentials and introduce the complex quantity $F_n$.

\begin{gather}
f(x) = \sum_{n = -\infty}^{\infty} F_n e^{i k_n x}
\end{gather}

Calculations yeild the following.

\begin{gather}
F_n = \frac{1}{L} \int_0^L f(x) e^{-i k_n x} dx
\end{gather}

Note that

\begin{gather}
F^*_n = F_{-n}
\end{gather}

Connect this to the old notation.

\begin{gather}
A_n = \frac{1}{2} (F_n + F^*_n) \qquad
B_n = \frac{i}{2} (F_n - F^*_n)
\end{gather}

#### Delta functions ####

Consistency leads to the following.

\begin{gather}
\delta(x - x') = \frac{1}{L} \sum_{n = -\infty}^{\infty} e^{i k_n (x - x')} \\
\delta_{n n'} = \frac{1}{L} \int_0^L e^{-i (k_n - k_{n'}) x} dx
\end{gather}

#### Complex functions ####

Consider a complex function $w(x)$ between $0$ and $L$. 

\begin{gather}
w(x) = u(x) + i v(x)
\end{gather}

Define.

\begin{gather}
u(x) = \sum_{n = -\infty}^{\infty} U_n e^{i k_n x} \qquad
U_n = \frac{1}{L} \int_0^L u(x) e^{-i k_n x} dx \\
v(x) = \sum_{n = -\infty}^{\infty} V_n e^{i k_n x} \qquad
V_n = \frac{1}{L} \int_0^L v(x) e^{-i k_n x} dx \\
w(x) = \sum_{n = -\infty}^{\infty} W_n e^{i k_n x} \qquad
W_n = \frac{1}{L} \int_0^L w(x) e^{-i k_n x} dx
\end{gather}

Calculations yeild.

\begin{gather}
U_n = \frac{1}{2}(W_n + W^*_{-n}) \\
V_n = -\frac{i}{2}(W_n - W^*_{-n})
\end{gather}

Hence, we can store two real functions in the complex quantity $W_n$.

### Fourier transform: Discrete 2D ###

Consider a real function $f(x, y)$ in an $L$ by $L$ square.  Start as before.

\begin{gather}
f(x, y) = \sum_{n = -\infty}^{\infty} F_n(y) e^{ i k_n x} \\
F_n(y) = \frac{1}{L} \int_0^L f(x, y) e^{-i k_n x} dx
\end{gather}

Now, Foureir expand the complex quantity $F_n(y)$.

\begin{gather}
F_n(y) = \sum_{m = -\infty}^{\infty} F_{nm} e^{i k_m y} \\
F_{nm} = \frac{1}{L} \int_0^L F_n(y) e^{- i k_m y} dy
\end{gather}

Plug into the previous equations.

\begin{gather}
f(x, y) = \sum_{n = -\infty}^{\infty} \sum_{m = -\infty}^{\infty} F_{nm} e^{i (k_n x + k_m y)} \\
F_{nm} = \left(\frac{1}{L}\right)^2 \int_0^L \int_0^L f(x, y) e^{-i \left(k_n x + k_m y\right)} dx dy
\end{gather}

The generalization to higher dimensions is clear.

### Fourier Transform: Continuous 1D ###

Consider a real function $f(x)$ on the entire real line.  In this case, a different starting notation is better.  First, shift the domain.

\begin{gather}
f(x) = \sum_{n = -\infty}^{\infty} F_n e^{i k_n x} \\
F_n = \frac{1}{L} \int_{-L/2}^{L/2} f(x) e^{-i k_n x} dx
\end{gather}

Next, define $\tilde f(k_n) = F_n L$ and $\Delta k_n = 2 \pi / L$.

\begin{gather}
f(x) = \frac{1}{2 \pi} \sum_{n = -\infty}^{\infty} \tilde f(k_n) e^{  i k_n x} \Delta k_n \\
\tilde f(k_n) = \int_{-L/2}^{L/2} f(x) e^{- i k_n x} dx
\end{gather}

Finally, consider the limit $L \rightarrow \infty$ where $k_n \rightarrow k$. 

\begin{gather}
f(x) = \frac{1}{2 \pi} \int_{-\infty}^{\infty} \tilde f(k) e^{  i k x} dk \\
\tilde f(k) = \int_{-\infty}^{\infty} f(x) e^{- i k x} dx
\end{gather}

#### Delta functions ####


Consistency leads to the following.

\begin{gather}
\delta(x - x') = \frac{1}{2 \pi} \int_{-\infty}^{\infty} e^{i k (x - x')} dk \\
\delta(k - k') = \frac{1}{2 \pi} \int_{-\infty}^{\infty} e^{- i (k - k') x} dx
\end{gather}

The above suggests that $\delta(-x) = \delta(x)$.  In general, it can be shown that

\begin{gather}
\delta(a x) = \frac{1}{\lvert a \lvert} \delta(x)
\end{gather}



### Towards Green's Functions ###

Change the context to position and time.  Consider a driven simple harmonic oscillator.

\begin{gather}
\frac{d^2 x}{d t^2} + \omega_0^2 x = \frac{1}{m} f(t)
\end{gather}

#### Continuous case ####

From before.

\begin{gather}
x(t) = \frac{1}{2 \pi} \int_{-\infty}^{\infty} \tilde x(\omega) e^{i \omega t} d\omega \qquad
\tilde x(\omega) = \int_{-\infty}^{\infty} x(t) e^{- i \omega t} dt
\end{gather}

\begin{gather}
\delta(t - t') = \frac{1}{2 \pi} \int_{-\infty}^{\infty} e^{i \omega (t - t')} d\omega \qquad
\delta(\omega - \omega') = \frac{1}{2 \pi} \int_{-\infty}^{\infty} e^{- i (\omega - \omega') t} dt
\end{gather}

Substitue equantion $x(t)$ into equation the differential equation.

\begin{split}
\frac{1}{2 \pi} \int_{-\infty}^\infty (\omega_0^2 - \omega^2) \tilde x(\omega) e^{i \omega t} d\omega
= \frac{1}{m} f(t)
\end{split}

Solve for $\tilde x(\omega)$.  Multiplying by $e^{-i \omega' t}$ and integrate with respect to $t$.  Note the delta function.  Change from $\omega'$ to $\omega$ and $t$ to $t'$.

\begin{gather}
\tilde x(\omega) = \int_{-\infty}^\infty \frac{1}{\omega_0^2 - \omega^2} e^{-i \omega t'} \frac{1}{m} f(t') dt'
\end{gather}

Solve for $x(t)$.  Multiply by $e^{i \omega t}/2\pi$.  Integrate with respect to $\omega$.

\begin{gather}
x(t) = \int_{-\infty}^\infty \frac{1}{2 \pi} 
\int_{-\infty}^\infty \frac{1}{\omega_0^2 - \omega^2} e^{i \omega (t - t')}  d\omega \frac{1}{m} f(t') dt'
\end{gather}

Define the Green's function.

\begin{gather}
x(t) = \int_{-\infty}^\infty G(t, t') \frac{1}{m} f(t') dt' \qquad
G(t, t') = \frac{1}{2 \pi} 
\int_{-\infty}^\infty \frac{1}{\omega_0^2 - \omega^2} e^{i \omega (t - t')} d\omega
\end{gather}

One can show that $G(t, t') = G(t', t)$.  In addition, one can show

\begin{gather}
\bigg[ \frac{d^2}{d t^2} + \omega_0^2 \bigg] G(t, t') = \delta(t - t')
\end{gather}

#### Discrete case ####

From before.

\begin{gather}
x(t) = \sum_{n = -\infty}^{\infty} X_n e^{i \omega_n t} \qquad
X_n = \frac{1}{T} \int_0^T x(t) e^{-i \omega_n t} dt
\end{gather}

\begin{gather}
\delta(t - t') = \frac{1}{T} \sum_{n = -\infty}^{\infty} e^{i \omega_n (t - t')} \qquad
\delta_{n n'} = \frac{1}{T} \int_0^T e^{-i (\omega_n - \omega_{n'}) t} dt
\end{gather}


Substitue equantion $x(t)$ into equation the differential equation.

\begin{split}
\sum_{n = -\infty}^{\infty} (\omega_0 - \omega_n) X_n e^{i \omega_n t} 
= \frac{1}{m} f(t)
\end{split}

Solve for $X_n$.  Multiplying by $e^{-i \omega'_n t}$ and integrate with respect to $t$.  Note the delta function.  Change from $\omega'_n$ to $\omega_n$ and $t$ to $t'$.

\begin{split}
X_n = \frac{1}{T} \int_0^T \frac{1}{\omega_0^2 - \omega_n^2} e^{-i \omega_n t'} \frac{1}{m} f(t) dt'
\end{split}

Solve for $x(t)$.  Multiply by $e^{i \omega_n t}$.  Sum with respect to $n$.

\begin{split}
x(t) = \int_0^T \frac{1}{T} \sum_{n = -\infty}^{\infty} \frac{1}{\omega_0^2 - \omega_n^2} e^{i \omega_n (t - t')} \frac{1}{m} f(t) dt'
\end{split}

Define the Green's function.

\begin{gather}
x(t) = \int_0^T G(t, t') \frac{1}{m} f(t') dt' \qquad
G(t, t') = \frac{1}{T} \sum_{n = -\infty}^{\infty} \frac{1}{\omega_0^2 - \omega_n^2} e^{i \omega_n (t - t')}
\end{gather}

One can show that $G(t, t') = G(t', t)$.  In addition, one can show

\begin{gather}
\frac{d^2}{d t^2} G(t, t') + \omega_0^2 G(t, t') = \delta(t - t')
\end{gather}
