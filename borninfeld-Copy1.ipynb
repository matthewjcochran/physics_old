### Maxwell's Equations in Tensor Form

A good way to start thinking about tensor is thought the continuity equation.  Let us introduce the the four-derivative and four-current.

\begin{gather}
\partial_\alpha
= \begin{pmatrix} \frac{1}{c}\partial_t & \partial_x & \partial_y & \partial_z \end{pmatrix} \qquad
J^\alpha
= \begin{pmatrix} c\rho & J_x & J_y & J_z \end{pmatrix}  
\end{gather}

The continuity equation can be written in a compact form.

\begin{gather}
\partial_\alpha J^\alpha = 0
\quad\Rightarrow\quad
\frac{\partial \rho}{\partial t} + \nabla \cdot J = 0
\end{gather}




Introduced the electromagnetic tensor and the electromagnetic pseudo tensor.  The first index represents the row and the second represents the column.

\begin{gather}
F^{\alpha \beta} = \begin{pmatrix}
0       & -E_x / c & -E_y / c & -E_z / c \\
E_x / c &  0       & -B_z     &  B_y     \\
E_y / c &  B_z     &    0     & -B_x     \\
E_z / c & -B_y     &  B_x     &  0       \end{pmatrix}
\end{gather}




We rewrite Maxwell's equations in preparation for tensor notation.  We note that $\epsilon_0 \mu_0 = \frac{1}{c^2}$ and collect the constant $c$ in front of the time partial and in front of the charge density.

\begin{gather}
\nabla \cdot \frac{E}{c} = \mu_0 c \rho \\
- \frac{1}{c} \frac{\partial}{\partial t} \frac{E}{c} + \nabla \times B = \mu_0 J \\
\nabla \cdot  B = 0 \vphantom{\frac{\partial B}{\partial t}} \\
- \frac{1}{c} \frac{\partial B}{\partial t} - \nabla \times \frac{E}{c} = 0
\end{gather}

This suggests have introduced the electromagnetic tensor along with the current tensor.  The first index represents the row and the second represents the column.

\begin{gather}
F^{\alpha \beta} = \begin{pmatrix}
0       & -E_x / c & -E_y / c & -E_z / c \\
E_x / c &  0       & -B_z     &  B_y     \\
E_y / c &  B_z     &    0     & -B_x     \\
E_z / c & -B_y     &  B_x     &  0       \end{pmatrix}
\end{gather}

Now we can rewrite the first two Maxwell's equations.

\begin{equation}
\partial_\alpha F^{\alpha \beta} = \mu_0 J^\beta
\quad \Rightarrow \quad
\begin{cases}
\nabla \cdot \dfrac{E}{c} = \mu_0 c \rho \\
- \dfrac{1}{c} \dfrac{\partial}{\partial t} \dfrac{E}{c} + \nabla \times B = \mu_0 J
\end{cases}
\end{equation}

There is a geometric similarity to the first two where the current is zero and $E/c \rightarrow B$ and $B \rightarrow - E/c$.  This suggests that we introduce the pseudomatrix.

\begin{gather}
G^{\alpha \beta} = \begin{pmatrix}
0   & -B_x     & -B_y     & -B_z     \\
B_x &  0       &  E_z / c & -E_y / c \\
B_y & -E_z / c &        0 &  E_x / c \\
B_z &  E_y / c & -E_x / c &  0
\end{pmatrix}
\end{gather}

We can rewrite the second two Maxwell equations.  

\begin{equation}
\partial_\alpha G^{\alpha \beta} = 0
\quad \Rightarrow \quad
\begin{cases}
\nabla \cdot  B = 0 \vphantom{\dfrac{\partial B}{\partial t}} \\
- \dfrac{1}{c} \dfrac{\partial B}{\partial t} - \nabla \times \dfrac{E}{c} = 0
\end{cases}
\end{equation}


<h2>Maxwell's Equations</h2>

We will start by reviewing Maxwell's equations in SI units.  Later, we will continue to use SI units.  This will involved keeping track of a few constants.  We hope that that in the long run this will allow for better insight.

\begin{gather}
\nabla \cdot E = \frac{\rho}{\epsilon_0} \\
\nabla \times B = \mu_0 J + \epsilon_0 \mu_0 \frac{\partial E}{\partial t} \\
\nabla \cdot B = 0 \vphantom{\frac{\partial B}{\partial t}} \\
\nabla \times E = - \frac{\partial B}{\partial t}
\end{gather}

The second two equations can be rewritten using in terms of potentials.

\begin{gather}
B = \nabla \times A \vphantom{\frac{\partial A}{\partial t}} \\
E = -\nabla \phi - \frac{\partial A}{\partial t}
\end{gather}

The power density and Lorentz force density law.

\begin{gather}
p = E \cdot J \\
f = \rho E + J \times B
\end{gather}

<h2>General Lagrangian</h2>

In general, we have $\mathcal{L}(\partial_\alpha A_\beta, A_\beta)$.  This notation is a little different and assumes that the Lagrangian contains all values 0 to 3. This gives the equation.

\begin{gather}
\partial_\alpha \frac{\partial \mathcal{L}}{\partial (\partial_\alpha A_\beta)}
= \frac{\partial \mathcal{L}}{\partial A_\beta}
\end{gather}

In many cases like the above, we have $\mathcal{L}(F_{\alpha\beta}, A_\beta)$.

\begin{equation}
\begin{split}
\frac{\partial \mathcal{L}}{\partial(\partial_\alpha A_\beta)}
&=\frac{\partial \mathcal{L}}{\partial F_{\mu\nu}}\frac{\partial F_{\mu\nu}}{\partial(\partial_\alpha A_\beta)}\\
&=\frac{\partial \mathcal{L}}{\partial F_{\mu\nu}}\frac{\partial}{\partial(\partial_\alpha A_\beta)}
  (\partial_\mu A_\nu - \partial_\nu A_\mu)\\
&=\frac{\partial \mathcal{L}}{\partial F_{\mu\nu}}
  (\delta^\alpha_\mu \delta^\beta_\nu - \delta^\alpha_\nu \delta^\beta_\mu)\\
&=\frac{\partial \mathcal{L}}{\partial F_{\alpha\beta}}
 -\frac{\partial \mathcal{L}}{\partial F_{\beta\alpha}}\\
&=2\frac{\partial \mathcal{L}}{\partial F_{\alpha\beta}}
\end{split}
\end{equation}

This allows us to write.

\begin{gather}
\partial_\alpha 2 \frac{\partial \mathcal{L}}{\partial F_{\alpha\beta}}
= \frac{\partial \mathcal{L}}{\partial A_\beta}
\end{gather}

Let us go to four space.

\begin{gather}
\frac{\partial\mathcal{L}}{\partial E_x}
= \frac{\partial\mathcal{L}}{\partial F_{0 1}} \frac{\partial F_{0 1}}{\partial E_x}
+ \frac{\partial\mathcal{L}}{\partial F_{1 0}} \frac{\partial F_{1 0}}{\partial E_x}
= \frac{2}{c}\frac{\partial\mathcal{L}}{\partial F_{0 1}}
\qquad
\frac{\partial\mathcal{L}}{\partial B_x}
= \frac{\partial\mathcal{L}}{\partial F_{3 2}} \frac{\partial F_{3 2}}{\partial B_x}
+ \frac{\partial\mathcal{L}}{\partial F_{2 3}} \frac{\partial F_{2 3}}{\partial B_x}
= 2 \frac{\partial\mathcal{L}}{\partial F_{3 2}}
\\
\frac{\partial\mathcal{L}}{\partial E_y}
= \frac{\partial\mathcal{L}}{\partial F_{0 2}} \frac{\partial F_{0 2}}{\partial E_y}
+ \frac{\partial\mathcal{L}}{\partial F_{2 0}} \frac{\partial F_{2 0}}{\partial E_y}
= \frac{2}{c}\frac{\partial\mathcal{L}}{\partial F_{0 2}}
\qquad
\frac{\partial\mathcal{L}}{\partial B_y}
= \frac{\partial\mathcal{L}}{\partial F_{1 3}} \frac{\partial F_{1 3}}{\partial B_y}
+ \frac{\partial\mathcal{L}}{\partial F_{3 1}} \frac{\partial F_{3 1}}{\partial B_y}
= 2 \frac{\partial\mathcal{L}}{\partial F_{1 3}}
\\
\frac{\partial\mathcal{L}}{\partial E_z}
= \frac{\partial\mathcal{L}}{\partial F_{0 3}} \frac{\partial F_{0 3}}{\partial E_z}
+ \frac{\partial\mathcal{L}}{\partial F_{3 0}} \frac{\partial F_{3 0}}{\partial E_z}
= \frac{2}{c}\frac{\partial\mathcal{L}}{\partial F_{0 3}}
\qquad
\frac{\partial\mathcal{L}}{\partial B_z}
= \frac{\partial\mathcal{L}}{\partial F_{2 1}} \frac{\partial F_{2 1}}{\partial B_z}
+ \frac{\partial\mathcal{L}}{\partial F_{1 2}} \frac{\partial F_{1 2}}{\partial B_z}
= 2 \frac{\partial\mathcal{L}}{\partial F_{2 1}}
\end{gather}

Using these chair rule results.
