### Maxwell's Equations ###

Let us define the following four-vectors and tensors.

\begin{gather}
\partial_\alpha = \begin{pmatrix} \frac{1}{c}\partial_t & \partial_x & \partial_y & \partial_z \end{pmatrix}
\qquad
A^\alpha = \left( \begin{matrix}\frac{1}{c}\phi & A_x & A_y & A_z \end{matrix}\right)
\qquad
J^\alpha = \begin{pmatrix} c\rho & J_x & J_y & J_z \end{pmatrix}  
\end{gather}

\begin{gather}
F^{\alpha \beta} = \begin{pmatrix}
0  & -\tfrac{1}{c} E_x & -\tfrac{1}{c} E_y & -\tfrac{1}{c} E_z \\
\tfrac{1}{c} E_x &  0   & -B_z &  B_y     \\
\tfrac{1}{c} E_y &  B_z &    0 & -B_x     \\
\tfrac{1}{c} E_z & -B_y &  B_x &  0 \end{pmatrix}
\qquad
P^{\alpha\beta} = \begin{pmatrix}
0 & -c D_x & -c D_y & -c D_z \vphantom{\tfrac{1}{c}} \\
 c D_x & 0 & -H_z & H_y \vphantom{\tfrac{1}{c}} \\
 c D_y & H_z & 0 & -H_x \vphantom{\tfrac{1}{c}} \\
 c D_z & -H_y & H_x & 0 \vphantom{\tfrac{1}{c}}
\end{pmatrix}
\end{gather}

With these, Maxwell's equations can be written in a compact form.

\begin{gather}
F^{\alpha \beta} = \partial^\alpha A^\beta - \partial^\beta A^\alpha
\quad \Rightarrow \quad
\begin{cases}
E = -\nabla \phi - \dfrac{\partial A}{\partial t} \\
B = \nabla \times A \vphantom{\dfrac{\partial A}{\partial t}}
\end{cases}
\end{gather}

The first pair of equations can be viewed as the definition of the electric and magnetic fields.  As we will see, the second pair of equations can be derived from a general Lagrangian. 

In general, we have $\mathcal{L}(\partial_\alpha A_\beta, A_\beta)$.  This notation is a different.  The Lagrangian in not a function of $\partial_\alpha A_\beta$ and $A_\beta$ for particular values of $\alpha$ and $\beta$.  Rather it is a functions of all combinations of values between 0 and 3. That is, the Lagrangian is a function of twenty variables.  Minimumizing the action gives

\begin{gather}
\partial_\alpha \frac{\partial \mathcal{L}}{\partial (\partial_\alpha A_\beta)}
= \frac{\partial \mathcal{L}}{\partial A_\beta}
\end{gather}

For the particular case where $\mathcal{L}(F_{\alpha\beta}, A_\beta)$, one can show that the previuos equation becomes

\begin{gather}
\partial_\alpha 2 \frac{\partial \mathcal{L}}{\partial F_{\alpha\beta}}
= \frac{\partial \mathcal{L}}{\partial A_\beta}
\end{gather}

With the following definitons
\begin{gather}
P^{\alpha\beta} = -2 \frac{\partial \mathcal{L}}{\partial F_{\alpha\beta}}
\quad \Rightarrow \quad
\begin{cases}
D = \nabla_E \mathcal{L} \vphantom{\dfrac{\partial A}{\partial t}} \\
H = -\nabla_B \mathcal{L} \vphantom{\dfrac{\partial A}{\partial t}}
\end{cases}
\end{gather}

\begin{gather}
J^\beta = -\frac{\partial \mathcal{L}}{\partial A_\beta}
\quad \Rightarrow \quad
\begin{cases}
\rho = - \dfrac{\partial \mathcal{L}}{\partial \phi} \vphantom{\dfrac{\partial A}{\partial t}} \\
J = \nabla_A \mathcal{L} \vphantom{\dfrac{\partial A}{\partial t}}
\end{cases}
\end{gather}

the previous equation becomes

\begin{gather}
\partial_\alpha P^{\alpha \beta} = J^\beta
\quad \Rightarrow \quad
\begin{cases}
\nabla \cdot D = \rho \vphantom{\dfrac{\partial A}{\partial t}} \\
- \dfrac{\partial D}{\partial t} + \nabla \times H = J
\end{cases}
\end{gather}

We see that the second pair of Maxwell's equations can be seen as arising from a general Lagrangian.  In general, $D$ and $H$ are complicated functions of $E$ and $B$.

### The Energy-Momentum Tensor ###

Gerardo Muñoz "Lagrangian field theories and energy‐momentum tensors". American Journal of Physics. 64, 1153 (1996)

\begin{split}
\Theta^{\alpha}_{\phantom\alpha \beta}
&= \frac{\partial \mathcal{L}}{\partial(\partial_\alpha A_\gamma)}
(\partial_\beta A_\gamma - \partial_\gamma A_\beta)
- \delta^\alpha_{\phantom\alpha \beta} \mathcal{L}
+ \frac{\partial \mathcal{L}}{\partial A_\alpha} A_\beta \\
&= 2 \frac{\partial \mathcal{L}}{\partial F_{\alpha \gamma}}
F_{\beta \gamma} - \delta^\alpha_{\phantom\alpha \beta} \mathcal{L}
+ \frac{\partial \mathcal{L}}{\partial A_\alpha} A_\beta \\
&= - P^{\alpha \gamma} F_{\beta \gamma} - \delta^\alpha_{\phantom\alpha \beta} \mathcal{L}
-J^\alpha A_\beta
\vphantom{\dfrac{\partial A}{\partial t}}
\end{split}

With the following definitions

\begin{gather}
u = D \cdot E - \mathcal{L} \qquad
S = c^2 D \times B \qquad
g = \frac{1}{c^2} E \times H \\
T_{i j} = D_i E_j + B_i H_j - (B \cdot H + \mathcal{L}) \delta_{i j}
\end{gather}

we get the following

\begin{gather}
\Theta^{\alpha\beta} =
\begin{pmatrix}
u & \frac{1}{c} S_x & \frac{1}{c} S_y & \frac{1}{c} S_z \\
c g_x & -T_{x x} & -T_{x y} & -T_{x z} \vphantom{\frac{1}{c} J_x \phi} \\
c g_y & -T_{y x} & -T_{y y} & -T_{z y} \vphantom{\frac{1}{c} J_x \phi} \\
c g_z & -T_{z x} & -T_{z y} & -T_{z z} \vphantom{\frac{1}{c} J_x \phi}
\end{pmatrix}
-
\begin{pmatrix}
\rho \phi & c \rho A_x & c \rho A_y & c \rho A_z \vphantom{\frac{1}{c} J_x \phi} \\
\frac{1}{c} J_x \phi & -J_x A_x & -J_x A_y & -J_x A_z \\
\frac{1}{c} J_y \phi & -J_y A_x & -J_y A_y & -J_z A_y \\
\frac{1}{c} J_z \phi & -J_z A_x & -J_z A_y & -J_z A_z
\end{pmatrix}
\end{gather}

Let us consider the Hamiltonian.

\begin{split}
\mathcal H
&= \dfrac{\partial \mathcal L}{\partial F_{\alpha \beta}} F^{\alpha \beta} - \mathcal L \\
&= -\dfrac{1}{2} P^{\alpha \beta} F_{\alpha \beta} - \mathcal L \\
&= D \cdot E - \mathcal L \vphantom{\dfrac{1}{2}} \\
\end{split}

We see that it is the same as the energy density, $u$.

With this convention we are allowed to raise and lower indicies as before.  Or, thinking in another way, there is always one 0 so by raising or lowering we cange the sign.

\qquad
F_{\alpha\beta} = -\frac{1}{2} \epsilon_{\alpha\beta\mu\nu} G^{\mu\nu}
\end{gather}

\begin{gather}
G^{\alpha \beta} = \begin{pmatrix}
0 & -B_x & -B_y & -B_z \vphantom{\tfrac{1}{c}} \\

B_y & -\tfrac{1}{c} E_z & 0 & \tfrac{1}{c} E_x \\
B_z & \tfrac{1}{c} E_y & -\tfrac{1}{c} E_x & 0
\end{pmatrix}
\end{gather}

Let us define the invariants.

\begin{gather}
\mathcal F = -\frac{1}{4 \mu_0} F_{\mu\nu} F^{\mu\nu}
= \frac{1}{2 \mu_0} \left( \frac{E^2}{c^2} - B^2 \right) \\
\mathcal G = -\frac{1}{4 \mu_0} F_{\mu\nu} G^{\mu\nu}
= \frac{1}{\mu_0 c} E \cdot B
\end{gather}

Let us assume $\mathcal L (\mathcal F, \mathcal G)$

\begin{gather}
P^{\alpha \beta} =
\frac{1}{\mu_0} \left(
\frac{\partial \mathcal L}{\partial \mathcal F} F^{\alpha \beta} + \frac{\partial \mathcal L}{\partial \mathcal G} G^{\alpha \beta} \right)
\quad \Rightarrow \quad
\begin{cases}
D = \epsilon_0 \left(
\dfrac{\partial \mathcal L}{\partial \mathcal F} E
+ \dfrac{\partial \mathcal L}{\partial \mathcal G} c B \right) \\
H = \dfrac{1}{\mu_0} \left(
\dfrac{\partial \mathcal L}{\partial \mathcal F} B
- \dfrac{\partial \mathcal L}{\partial \mathcal G} \dfrac{E}{c} \right)
\end{cases}
\end{gather}

\begin{gather}
u = \frac{\partial \mathcal L}{\partial \mathcal F} \epsilon_0 E^2
+ \frac{\partial \mathcal L}{\partial \mathcal G} \mathcal G
- \mathcal L \\

T_{i j} = \frac{\partial \mathcal L}{\partial \mathcal F} \left(
\epsilon_0 E_i E_j + \frac{1}{\mu_0} B_i B_j \right)
