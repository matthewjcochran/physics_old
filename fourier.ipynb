<h2>Fourier transform: Discrete 1D</h2>

<p>Consider a real function $f(x)$ that is periodic between $0$ and $L$.</p>

<h3>The base defintion</h3>

<p>The first definition has all real functions and positive indicies.

$$ f(x) = A_0 + \sum_{n = 1}^{N} \left[
          A_n \cos \left(\frac{2 \pi n}{L} x\right)
        + B_n \sin \left(\frac{2 \pi n}{L} x\right) 
   \right] $$

$$ A_0 = \frac{1}{L} \int_0^L f(x) dx $$

$$ A_n = \frac{2}{L} \int_0^L f(x) \cos \left(\frac{2 \pi n}{L} x\right) dx $$

$$ B_n = \frac{2}{L} \int_0^L f(x) \sin \left(\frac{2 \pi n}{L} x\right) dx $$</p>

<h3>The first notational update</h3>

<p>We can compacitfy things up by allowing $n$ to be negative and setting $k_n = 2 \pi n / L$.  

$$ f(x) = \sum_{n = -N}^{N} \left[
          a_n \cos \left(k_n x\right)
        + b_n \sin \left(k_n x\right) 
   \right] $$

$$ a_n = \frac{1}{L} \int_0^L f(x) \cos \left(k_n x\right) dx $$

$$ b_n = \frac{1}{L} \int_0^L f(x) \sin \left(k_n x\right) dx $$

We can connect this to the older notation.

$$ A_0 = a_0 $$

$$ A_n = 2 a_n, \quad n > 1 $$

$$ B_n = 2 b_n, \quad n > 1 $$

Also, note the following.

$$ a_{-n} = a_n $$

$$ b_{-n} = - b_n $$

</p>

<h3>The second notational update</h3>

<p>Let us write the trig functions as complex exponentials and introduce new variables.

$$ c_n    = a_n - i b_n $$

$$ c_{-n} = a_n + i b_n $$


This allows us to write the following.

$$ f(x) = \sum_{n = -N}^{N} c_n    e^{  i k_n x} $$

$$ c_n = \frac{1}{L} \int_0^L f(x) e^{- i k_n x} dx $$

We can connect to the origionla notation.

$$ A_0 = c_0 $$

$$ A_n = c_n + c_{-n}, \quad n > 1 $$

$$ B_n = i (c_n - c_{-n}), \quad n > 1 $$

</p>

<h3>Delta functions</h3>

<p>Using a circular arguemtents we discover delta functions.  The first delta function is implied by the following.

$$ f(x) = \int_0^L f(x') \frac{1}{L} \sum_{n = -\infty}^{\infty} e^{i k_n (x - x')} dx' $$

$$ \delta(x - x') = \frac{1}{L} \sum_{n = -\infty}^{\infty} e^{i k_n (x - x')} $$

The second delta functions is implied in a similar way.

$$ c_n = \sum_{n' = -\infty}^{\infty} c_{n'} \frac{1}{L} \int_0^L e^{-i (k_n - k_{n'}) x} dx $$

$$ \delta_{n n'} = \frac{1}{L} \int_0^L e^{-i (k_n - k_{n'}) x} dx $$

</p>

<h2>Fourier transform: Discrete 2D</h2>

<p>Consider a two dimentional functions.  We can start as before.

$$ f(x, y) = \sum_{n = -N}^{N} c_n(y)    e^{  i k_n x} $$

$$ c_n(y) = \frac{1}{L} \int_0^L f(x, y) e^{- i k_n x} dx $$

Now, Foureir expand $c_n(y)$.

$$ c_n(y) = \sum_{m = -M}^{M} c_{nm} e^{i k_m y} $$

$$ c_{nm} = \frac{1}{L} \int_0^L c_n(y) e^{- i k_m y} dy $$

Plugging back into the first equations we have the following symmetric relationship.

$$ f(x, y) = \sum_{n = -N}^{N} \sum_{m = -M}^{M} c_{nm} 
             e^{i (k_n x + k_m y)} $$

$$ c_{nm} = \left(\frac{1}{L}\right)^2
            \int_0^L \int_0^L f(x, y)
            e^{ -i \left(k_n x + k_m y\right)} dx dy $$
</p>

<h2>Fourier Transform: Continuous 1D</h2>

<p>Considering a continuous limit, we see that a different starting notation would have been better.  Hence, we introduce $c_n = \frac{1}{L} c(k_n)$.  With this change in notation, noting $\Delta k_n = \frac{2 \pi}{L}$, and a shift in the integration limits we have the following.

$$ f(x) = \frac{1}{2 \pi} \sum_{n = -\infty}^{\infty} c(k_n) e^{ i k_n x} \Delta k_n $$

$$ c(k_n) = \int_{-L/2}^{L/2} f(x) e^{- i k_n x} dx $$

Now, we take the limit as $L$ goes to $\infty$ as $k_n$ goes to $k$.

$$ f(x) = \frac{1}{2 \pi} \int_{-\infty}^{\infty} c(k) e^{ i k x} dk $$

$$ c(k) = \int_{-\infty}^{\infty} f(x) e^{- i k x} dx $$

</p>

<h3>The wave equation</h3>

<p> Consider the wave equation

$$ \frac{\partial^2 \phi}{\partial t^2} = c^2 \frac{\partial^2 \phi}{\partial x^2} $$

where

$$ \phi(x,0) = \phi_0(x) $$
$$ \dot\phi(x, 0) = \pi_0(x) $$

Let us take the Fourier transform of the equation.

$$ \int_{-\infty}^\infty e^{-i k x}
   \frac{\partial^2 \phi}{\partial t^2} dx
 = \int_{-\infty}^\infty e^{-i k x} c^2 \frac{\partial^2 \phi}{\partial x^2} dx $$

With some algebra, integration by parts, and assumptions about the function at infinity we get the following.

$$ \frac{\partial^2}{\partial t^2} \int_{-\infty}^\infty
   \phi(x, t) e^{-i k x} dx
 = -c^2 k^2 \int_{-\infty}^\infty \phi(x, t) e^{-i k x} dx $$

Now define the transform where there is a change in notation from before.

$$ \phi(x, t) = \frac{1}{2 \pi} \int_{-\infty}^\infty \hat\phi(k, t) e^{i k x} dk $$

$$ \hat\phi(k, t) = \int_{-\infty}^\infty \phi(x, t) e^{-i k x} dx $$

The wave equation then says

$$ \frac{d^2 \hat\phi}{d t^2} = -c^2 k^2 \hat\phi $$

This has the solution

$$ \hat\phi(k,t) = \hat f(k) e^{i c k t} + \hat g(k) e^{-i c k t} $$

so

$$ \phi(x, t) = \frac{1}{2 \pi} \int_{-\infty}^\infty \left[
   \hat f(k) e^{i k (x + c t)} + \hat g(k) e^{i k (x - c t)} \right] dk $$

This leads to a classical result, which we will not use later.

$$ \phi(x, t) = f(x + c t) + g(x - c t) $$

Now let us consider the initial conditions.  With some algebra you can show the following.

$$ \hat f(k) = \frac{1}{2} \left(\hat\phi_0(k) + \frac{1}{i c k}\hat\pi_0(k) \right) $$

$$ \hat g(k) = \frac{1}{2} \left(\hat\phi_0(k) - \frac{1}{i c k}\hat\pi_0(k) \right) $$

The resulting equation is fairly long so let us break it into two parts.

$$ \phi(x, t) = \phi_1(x, t) + \phi_2(x, t) $$

The first party is easy.

$$ \phi_1 (x, t) = \frac{1}{2 \pi} \int_{-\infty}^\infty \frac{1}{2} \left[
   \hat\phi_0(k) e^{i k (x + c t)}
 + \hat\phi_0(k) e^{i k (x - c t)} \right] dk $$

The second part is more complicated and involves some integation by parts.

$$ \phi_2(x, t) = \frac{1}{2 \pi} \int_{-\infty}^\infty \frac{1}{2} \left[
   \frac{1}{i c k}\hat\pi_0(k) e^{i k (x + c t)}
  -\frac{1}{i c k}\hat\pi_0(k) e^{i k (x - c t)}
   \right] dk $$
   
$$ \phi_2(x, t) = \frac{1}{2 \pi} \int_{-\infty}^\infty \frac{1}{2} 
   \int_0^{t}\left[
   \hat\pi_0(k) e^{i k (x + c t')}
  +\hat\pi_0(k) e^{i k (x - c t')}
   \right] dt' dk $$

The results are the follwoing.

$$ \phi_1 (x, t) = \frac{1}{2} \left[
   \phi_0(x + c t) + \phi_0(x - c t) \right] $$ 
   
$$ \phi_2(x, t) = \frac{1}{2} \int_0^t \left[
   \pi_0(x + c t') + \pi_0(x - c t') \right] dt' $$

Finally, we arrive at the d'Alembera solution by rewriting the second piece.

$$ \phi_2(x, t) = \frac{1}{2 c}
   \int_{x - c t}^{x + c t} \pi_0(\xi) d\xi $$

Here is the classical result.

$$ \phi(x, t) = \frac{1}{2} \left[
   \phi_0(x + c t) + \phi_0(x - c t) \right]
 + \frac{1}{2 c}
   \int_{x - c t}^{x + c t} \pi_0(\xi) d\xi $$
 
</p>

<h2>Wave Equation: Discrete 2D</h2>

$$ f(x, y, t) = \sum_{n = -N}^{N} \sum_{m = -M}^{M} c_{nm}(t) 
             e^{i (k_n x + k_m y)} $$

$$ c_{nm}(t) = \left(\frac{1}{L}\right)^2
            \int_0^L \int_0^L f(x, y, t)
            e^{ -i \left(k_n x + k_m y\right)} dx dy $$

$$ \frac{\partial^2 f}{\partial x^2} + \frac{\partial^2 f}{\partial y^2} - \frac{1}{c^2} \frac{\partial^2 f}{\partial t^2} = 0 $$

Plugging into the wave equation we get the following.

$$ \sum_{n = -N}^{N} \sum_{m = -M}^{M} \left[ 
-(k_n^2 + k_m^2) c_{nm} - \frac{1}{c^2} \frac{d^2 c_{nm}}{d t^2} \right] e^{i (k_n x + k_m y)} = 0 $$ 

Assuming periodic boundary conditions.

$$ \frac{d^2 c_{nm}}{d t^2} = -c^2 (k_n^2 + k_m^2) c_{nm} $$ 

$$ \frac{d^2 c_{nm}}{d t^2} = - \omega^2_{nm} c_{nm} $$

$$ c_{nm} = a_{nm} e^{i w_{nm} t} + b_{nm} e^{-i w_{nm} t} $$

$$ f(x, y, t) = \sum_{n = -N}^{N} \sum_{m = -M}^{M} (a_{nm} e^{i w_{nm} t} + b_{nm} e^{-i w_{nm} t}) e^{i (k_n x + k_m y)} $$


$$ f(x, y, t) = \sum_{n = -N}^{N} \sum_{m = -M}^{M}
  \left[ a_{nm} e^{i (k_n x + k_m y + w_{nm} t)}
 + b_{nm} e^{i (k_n x + k_m y - w_{nm} t)} \right] $$
 
$$ f_0 = \sum_{n = -N}^{N} \sum_{m = -M}^{M}
  \left[ a_{nm} + b_{nm} \right] e^{i (k_n x + k_m y)} $$

$$ \dot f_0 = \sum_{n = -N}^{N} \sum_{m = -M}^{M}
 i \omega_{nm} \left[ a_{nm} - b_{nm} \right] e^{i (k_n x + k_m y)}$$
 
$$ a_{nm} + b_{nm} 
 = \left(\frac{1}{L}\right)^2 \int_0^L \int_0^L
   f_0 (x, y) e^{-i (k_n x + k_m y)} dx dy $$

$$ a_{nm} - b_{nm} 
 = \left(\frac{1}{L}\right)^2 \int_0^L \int_0^L
   \frac{1}{i \omega_{nm}} \dot f_0 (x, y)
   e^{-i (k_n x + k_m y)} dx dy $$
   
$$ a_{nm} = \frac{1}{2}\left(\frac{1}{L}\right)^2 \int_0^L \int_0^L
   \left[ f_0 (x, y) + \frac{1}{i \omega_{nm}} \dot f_0 (x, y) \right]
   e^{-i (k_n x + k_m y)} dx dy $$   
   
$$ b_{nm} = \frac{1}{2}\left(\frac{1}{L}\right)^2 \int_0^L \int_0^L
   \left[ f_0 (x, y) - \frac{1}{i \omega_{nm}} \dot f_0 (x, y) \right]
   e^{-i (k_n x + k_m y)} dx dy $$

<h3>Rethink</h3>

$$ f(x, y, t) = \sum_{n = -N}^{N} \sum_{m = -M}^{M}
  \left[ a_{nm} e^{i (k_n x + k_m y + w_{nm} t)}
 + a^*_{nm} e^{-i (k_n x + k_m y + w_{nm} t)} \right] $$

$$ a_{nm} = \frac{1}{2}\left(\frac{1}{L}\right)^2 \int_0^L \int_0^L
   \left[ f_0 (x, y) + \frac{1}{i \omega_{nm}} \dot f_0 (x, y) \right]
   e^{-i (k_n x + k_m y)} dx dy $$   
   
$$ a^*_{nm} = \frac{1}{2}\left(\frac{1}{L}\right)^2 \int_0^L \int_0^L
   \left[ f_0 (x, y) - \frac{1}{i \omega_{nm}} \dot f_0 (x, y) \right]
   e^{i (k_n x + k_m y)} dx dy $$

<h3>Rethink Again</h3>

$$ f(x, y, t) = \sum_{n = -N}^{N} \sum_{m = -M}^{M}
  \left[ c_{nm} \cos(\omega_{nm} t)
  + \frac{\dot c_{nm}}{\omega_{nm}} \sin(\omega_{nm} t) 
  \right ]e^{i (k_n x + k_m y)} $$

$$ \omega_{nm} = c \sqrt{k_n^2 + k_m^2} $$

$$ c_{nm} = \left(\frac{1}{L}\right)^2
  \int_0^L \int_0^L f(x, y, 0)
  e^{ -i \left(k_n x + k_m y\right)} dx dy $$
  
$$ \dot c_{nm} = \left(\frac{1}{L}\right)^2
  \int_0^L \int_0^L \dot f(x, y, 0)
  e^{ -i \left(k_n x + k_m y\right)} dx dy $$

<h3>Simple Harmonic Oscillator</h3>

$$ m \frac{d^2 q}{d t^2} + k q = 0 $$

$$ q_0 = q(0) \qquad p_0 = m \dot q(0) $$

Has solution

$$ q(t) = q_0 \cos(\omega t)
  +  \frac{\dot q_0}{\omega} \sin (\omega t) $$
  
$$ q(t) = q_0 \frac{e^{i \omega t} + e^{-i \omega t}}{2}
  + \frac{\dot q_0}{\omega} \frac{e^{i \omega t} - e^{-i \omega t}}{2 i} $$
  
$$ q(t) = \frac{1}{2} \left(q_0 + \frac{\dot q_0}{i \omega} \right) e^{i \omega t}
  + \frac{1}{2} \left(q_0 - \frac{\dot q_0}{i \omega} \right) e^{- i \omega t} $$
  
$$ q(t) = \frac{1}{2} \left[a_0 e^{i \omega t} + a_0^* e^{-i \omega t} \right] $$
