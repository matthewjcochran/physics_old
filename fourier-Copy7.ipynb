### Fourier transform: Discrete 1D ###

Consider a real function $f(x)$ between $0$ and $L$.  We write this as a sum of periodic functions.

#### Base defintion ####

Start with real variables and positive indicies.

\begin{gather}
f(x) = a_0 + \sum_{n = 1}^{\infty}
\left[ a_n \cos \left(\frac{2 \pi n}{L} x\right) + b_n \sin \left(\frac{2 \pi n}{L} x\right) \right]
\end{gather}

Calculations yeild the following.

\begin{gather}
a_0 = \frac{1}{L} \int_0^L f(x) dx \\
a_n = \frac{2}{L} \int_0^L f(x) \cos \left(\frac{2 \pi n}{L} x\right) dx \\
b_n = \frac{2}{L} \int_0^L f(x) \sin \left(\frac{2 \pi n}{L} x\right) dx
\end{gather}

Note that

\begin{gather}
a_{-n} = a_n \qquad
b_{-n} = -b_n
\end{gather}

#### First notation update ####

Set $k_n = 2 \pi n / L$, allow $n$ to have integer values, and introduce $a_n$ and $b_n$.  

\begin{gather}
f(x) = \sum_{n = -\infty}^{\infty} \left[ A_n \cos \left(k_n x\right) + B_n \sin \left(k_n x\right) \right]
\end{gather}

Calculations yeild the following.

\begin{gather}
A_n = \frac{1}{L} \int_0^L f(x) \cos \left(k_n x\right) dx \\
B_n = \frac{1}{L} \int_0^L f(x) \sin \left(k_n x\right) dx
\end{gather}


Note that

\begin{gather}
A_{-n} = A_n \qquad
B_{-n} = -B_n
\end{gather}

Connect this to the old notation.

\begin{gather}
a_0 = A_0 \qquad
a_n = 2 A_n \qquad
b_n = 2 B_n
\end{gather}

#### Second notation update ####

Replace the tigonmetric functions with complex exponentials and introduce the complex quantity $F_n$.

\begin{gather}
f(x) = \sum_{n = -\infty}^{\infty} F_n e^{i k_n x}
\end{gather}

Calculations yeild the following.

\begin{gather}
F_n = \frac{1}{L} \int_0^L f(x) e^{-i k_n x} dx
\end{gather}

Note that

\begin{gather}
F^*_n = F_{-n}
\end{gather}

Connect this to the old notation.

\begin{gather}
A_n = \frac{1}{2} (F_n + F^*_n) \qquad
B_n = \frac{i}{2} (F_n - F^*_n)
\end{gather}

\begin{gather}
A_n = \Re(F_n) \qquad
B_n = -\Im(F_n)
\end{gather}

#### Delta functions ####

Consistency leads to the following.

\begin{gather}
\delta(x - x') = \frac{1}{L} \sum_{n = -\infty}^{\infty} e^{i k_n (x - x')} \\
\delta_{n n'} = \frac{1}{L} \int_0^L e^{-i (k_n - k_{n'}) x} dx
\end{gather}

One can show that $\delta(-x) = \delta(x)$ and that $\delta_{n' n} = \delta_{n n'}$.

#### Complex functions ####

Consider a complex function $w(x)$ between $0$ and $L$. 

\begin{gather}
w(x) = u(x) + i v(x)
\end{gather}

Define.

\begin{gather}
u(x) = \sum_{n = -\infty}^{\infty} U_n e^{i k_n x} \qquad
U_n = \frac{1}{L} \int_0^L u(x) e^{-i k_n x} dx \\
v(x) = \sum_{n = -\infty}^{\infty} V_n e^{i k_n x} \qquad
V_n = \frac{1}{L} \int_0^L v(x) e^{-i k_n x} dx \\
w(x) = \sum_{n = -\infty}^{\infty} W_n e^{i k_n x} \qquad
W_n = \frac{1}{L} \int_0^L w(x) e^{-i k_n x} dx
\end{gather}

Calculations yeild.

\begin{gather}
U_n = \frac{1}{2}(W_n + W^*_{-n}) \\
V_n = -\frac{i}{2}(W_n - W^*_{-n})
\end{gather}

Hence, we can store two real functions in the complex quantity $W_n$.

### Fourier transform: Discrete 2D ###

Consider a real function $f(x, y)$ in an $L$ by $L$ square.  Start as before.

\begin{gather}
f(x, y) = \sum_{m = -\infty}^{\infty} F_m(y) e^{ i k_m x} \\
F_m(y) = \frac{1}{L} \int_0^L f(x, y) e^{-i k_m x} dx
\end{gather}

Now, Foureir expand the complex quantity $F_m(y)$.

\begin{gather}
F_m(y) = \sum_{n = -\infty}^{\infty} F_{m n} e^{i k_n y} \\
F_{m n} = \frac{1}{L} \int_0^L F_m(y) e^{- i k_n y} dy
\end{gather}

Plug into the previous equations.

\begin{gather}
f(x, y) = \sum_{m = -\infty}^{\infty} \sum_{n = -\infty}^{\infty} F_{m n} e^{i (k_m x + k_n y)} \\
F_{m n} = \left(\frac{1}{L}\right)^2 \int_0^L \int_0^L f(x, y) e^{-i \left(k_m x + k_n y\right)} dx dy
\end{gather}

The generalization to higher dimensions is clear.

### Fourier Transform: Continuous 1D ###

Consider a real function $f(x)$ on the entire real line.  In this case, a different starting notation is better.  First, shift the domain.

\begin{gather}
f(x) = \sum_{n = -\infty}^{\infty} F_n e^{i k_n x} \\
F_n = \frac{1}{L} \int_{-L/2}^{L/2} f(x) e^{-i k_n x} dx
\end{gather}

Next, define $\tilde f(k_n) = F_n L$ and $\Delta k_n = 2 \pi / L$.

\begin{gather}
f(x) = \frac{1}{2 \pi} \sum_{n = -\infty}^{\infty} \tilde f(k_n) e^{  i k_n x} \Delta k_n \\
\tilde f(k_n) = \int_{-L/2}^{L/2} f(x) e^{- i k_n x} dx
\end{gather}

Finally, consider the limit $L \rightarrow \infty$ where $k_n \rightarrow k$. 

\begin{gather}
f(x) = \frac{1}{2 \pi} \int_{-\infty}^{\infty} \tilde f(k) e^{  i k x} dk \\
\tilde f(k) = \int_{-\infty}^{\infty} f(x) e^{- i k x} dx
\end{gather}

#### Delta functions ####


Consistency leads to the following.

\begin{gather}
\delta(x - x') = \frac{1}{2 \pi} \int_{-\infty}^{\infty} e^{i k (x - x')} dk \\
\delta(k - k') = \frac{1}{2 \pi} \int_{-\infty}^{\infty} e^{- i (k - k') x} dx
\end{gather}

The above suggests that $\delta(-x) = \delta(x)$.  In general, it can be shown that

\begin{gather}
\delta(a x) = \frac{1}{\lvert a \lvert} \delta(x)
\end{gather}



### Towards Green's Functions ###

Change to position and time.  Consider a driven simple harmonic oscillator.

\begin{gather}
\bigg( \frac{d^2}{d t^2} + \omega_0^2 \bigg) x(t) = \frac{1}{m} f(t)
\end{gather}

For the discrete case, substitue the expression for $x(t)$ into the differential equation.  Solve for $X_n$ by multiplying by $e^{-i \omega'_n t}$ and integrating with respect to $t$.  Note the delta function and change from $\omega'_n$ to $\omega_n$ and $t$ to $t'$.  Solve for $x(t)$ by multiply by $e^{i \omega_n t}/T$ and summing with respect to $n$.  The solution is written in terms of a Green's function.

\begin{gather}
x(t) = \int_0^T G(t, t') \frac{1}{m} f(t') dt' \\
G(t, t') = \frac{1}{T} \sum_{n = -\infty}^{\infty} \frac{1}{\omega_0^2 - \omega_n^2} e^{i \omega_n (t - t')}
\end{gather}

For the continuous case, substitue the expression for $x(t)$ into the differential equation and solve for $\tilde x(\omega)$ by multiplying by $e^{-i \omega' t}$ and integrating with respect to $t$.  Note the delta function and change from $\omega'$ to $\omega$ and $t$ to $t'$. Solve for $x(t)$ by multiply by $e^{i \omega t}/2\pi$ and integrating with respect to $\omega$.  The solution is written in terms of a Green's function.

\begin{gather}
x(t) = \int_{-\infty}^\infty G(t, t') \frac{1}{m} f(t') dt' \\
G(t, t') = \frac{1}{2 \pi} 
\int_{-\infty}^\infty \frac{1}{\omega_0^2 - \omega^2} e^{i \omega (t - t')} d\omega
\end{gather}

In both cases, $G(t, t') = G(t', t)$, $G^*(t, t') = G(t, t')$, and

\begin{gather}
\bigg( \frac{d^2}{d t^2} + \omega_0^2 \bigg) G(t, t') = \delta(t - t')
\end{gather}

#### More general Green's functions ####

Consider a linear differential operator $L$, which we assume is real.  Our goal is to find solutions to the equation

\begin{gather}
L x(t) = f(t)
\end{gather}

Assume that we can find some orthogonal solutions $u_n(t)$ to the homogeneous equation with eigenvalues $\lambda_n$.

\begin{gather}
L u_n(t) = \lambda_n u_n(t) \qquad
L u^*_n(t) = \lambda^*_n u^*_n(t)
\end{gather}

\begin{gather}
\delta(t - t') = \frac{1}{T} \sum_{n = -\infty}^{\infty} u_n(t) u^*_n(t') \\
\delta_{n n'} = \frac{1}{T} \int_0^T u_n(t) u^*_{n'}(t) dt
\end{gather}

Also assume

\begin{gather}
x(t) = \sum_{n = -\infty}^\infty X_n u_n(t)
\end{gather}

Calculations yield.

\begin{gather}
X_n = \frac{1}{T} \int_0^T x(t) u^*_n(t) dt
\end{gather}

\begin{gather}
x(t) = \int_0^T G(t, t') f(t') dt' \\
G(t, t') = \frac{1}{T} \sum_{n = -\infty}^{\infty} \frac{1}{\lambda_n} u(t)_n u^*_n(t')
\end{gather}

### Matrix albegra ###

Consider the square matrix $A$ and row vectors $x$ and $b$ that satisfy

\begin{gather}
\sum_n A_{m n} x_n = b_m
\qquad \Rightarrow \qquad
A x = b
\end{gather}

Define eigen values $\lambda^{(l)}$ and eigen vectors $u^{(l)}$ stored in rows such that

\begin{gather}
\sum_n A_{m n} u^{(l)}_n = \lambda^{(l)} u^{(l)}_m
\qquad \Rightarrow \qquad
A u^{(l)} = \lambda^{(l)} u^{(l)}
\end{gather}

\begin{gather}
\sum_n A^T_{m n} v^{(l)}_n = \lambda^{(l)} v^{(l)}_m
\qquad \Rightarrow \qquad
A^T v^{(l)} = \lambda^{(l)} v^{(l)}
\end{gather}

Define $U_{n l} = u^{(l)}_n$, $V_{n l} = v^{(l)}_n$, and $\Lambda_{n l} = \delta_{n l} \lambda^{(l)}$.

\begin{gather}
\sum_n A_{m n} U_{n l} = \sum_n U_{m n} \Lambda_{n l}
\qquad \Rightarrow \qquad
A U = U \Lambda
\end{gather}

\begin{gather}
\sum_n A^T_{m n} V_{n l} = \sum_n V_{m n} \Lambda_{n l}
\qquad \Rightarrow \qquad
A^T V = V \Lambda
\end{gather}

\begin{gather}
A U = U \Lambda
\qquad \Rightarrow \qquad
V^T A U = V^T U \Lambda
\end{gather}

\begin{gather}
V^T A = \Lambda V^T
\qquad \Rightarrow \qquad
V^T A U = \Lambda V^T U
\end{gather}

\begin{gather}
W = V^T U
\end{gather}

\begin{gather}
W \Lambda - \Lambda W = 0
\qquad \Rightarrow \qquad
(\lambda^{(m)} - \lambda^{(n)}) W_{m n} = 0
\end{gather}

\begin{gather}
W_{m n} = \delta_{m n} w^{(n)}
\end{gather}

\begin{gather}
U^T V = W
\qquad \Rightarrow \qquad
u^{(m)} \cdot v^{(n)} = \delta_{m n} w^{(n)}
\end{gather}

\begin{gather}
U^T V = V^T U = I
\end{gather}

\begin{gather}
U^{-1} = V^T \qquad
V^{-1} = U^T \qquad
(U^T)^{-1} = V \qquad
(V^T)^{-1} = U
\end{gather}

\begin{gather}
A = U \Lambda V^T \qquad
A^T = V \Lambda U^T
\end{gather}

\begin{gather}
(\Lambda^{-1})_{m n} = \delta_{m n} (\lambda^{-1})^{(n)}
\end{gather}

\begin{gather}
A^{-1} = U \Lambda^{-1} V^T \qquad
(A^T)^{-1} = V \Lambda^{-1} U^T
\end{gather}

\begin{gather}
A^{-1} U = U \Lambda^{-1} \qquad
(A^T)^{-1} V = V \Lambda^{-1}
\end{gather}
